````python
import numpy as np

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# 1. Initialize inputs, weights, and bias
x = np.array([1.5, 2.0, -1.0]) # Example input features
w = np.array([0.5, -0.2, 0.1]) # Learned weights
b = 0.1                        # Learned bias

# 2. Linear combination (Dot product)
z = np.dot(w, x) + b

# 3. Activation
y = sigmoid(z)
print(f"Neuron Output: {y:.4f}")
````

In the simplest terms, because we used the **Sigmoid activation function** in that example, the neuron's output ($y$) represents a **probability or a confidence score** between 0 and 1.

Here is a breakdown of what that means both conceptually and mathematically:

### 1. The Intuition: "Is this what I'm looking for?"
Think of a single neuron as a tiny detector trying to answer a simple "Yes or No" question.
* If the output is **close to 1** (e.g., 0.95), the neuron is "firing strongly." It is highly confident that it has found the specific pattern it was trained to look for.
* If the output is **close to 0** (e.g., 0.05), the neuron is inactive. It is confident that the pattern is *not* present.
* If the output is **around 0.5**, the neuron is confused or uncertain.

### 2. The Math Translation
To understand *how* it gets this score, we have to look at the two steps happening inside the neuron:

**Step A: The Dot Product ($z$)**
The linear calculation $z = \mathbf{w}^T\mathbf{x} + b$ is essentially measuring **alignment or similarity**.
* The weights ($\mathbf{w}$) represent the "ideal pattern" the neuron wants to see.
* The inputs ($\mathbf{x}$) are the actual data it is looking at right now.
* If the inputs align perfectly with the weights, $z$ will be a large positive number. If they are total opposites, $z$ will be a large negative number.

**Step B: The Activation ($y$)**
The problem is that $z$ can be any number from $-\infty$ to $+\infty$. That is hard to interpret. So, we pass it through the Sigmoid function: $y = g(z)$.
* A massive positive $z$ gets squished up to exactly **1.0**.
* A massive negative $z$ gets squished down to exactly **0.0**.
* A $z$ of exactly zero becomes exactly **0.5**.

**How does the alighment came?**
That is a great question. It is one thing to say the dot product measures "alignment," but seeing exactly *why* it does that makes the whole concept click.

The idea that the dot product measures alignment comes from linear algebra, and we can look at it in two ways: **Geometrically** (how we visualize it) and **Algebraically** (how the computer calculates it).

### 1. The Geometric View (The Angle)
Imagine the inputs ($\mathbf{x}$) and the weights ($\mathbf{w}$) are arrows (vectors) pointing in space.

There is a fundamental mathematical theorem that defines the dot product using the angle ($\theta$) between these two arrows:
$$\mathbf{w} \cdot \mathbf{x} = \|\mathbf{w}\| \|\mathbf{x}\| \cos(\theta)$$

Here, $\|\mathbf{w}\|$ and $\|\mathbf{x}\|$ are just the lengths of the arrows. The real magic happens with the **cosine of the angle** ($\cos(\theta)$).

* **Perfect Alignment ($\theta = 0^\circ$):** If the input vector points in the exact same direction as the weight vector, the angle is 0. Since $\cos(0^\circ) = 1$, the dot product is maximized. The neuron screams "YES!"
* **Unrelated / Orthogonal ($\theta = 90^\circ$):** If the vectors are perpendicular, they share nothing in common. Since $\cos(90^\circ) = 0$, the dot product is exactly 0. The neuron is completely indifferent.
* **Complete Opposition ($\theta = 180^\circ$):** If the vectors point in exact opposite directions, the angle is 180. Since $\cos(180^\circ) = -1$, the dot product becomes a large negative number. The neuron screams "NO!"



### 2. The Algebraic View (The Template Match)
While the geometric view is great for our brains, the computer calculates alignment algebraically by multiplying the numbers element by element and adding them up:
$$z = (w_1 \cdot x_1) + (w_2 \cdot x_2) + (w_3 \cdot x_3)$$

Think of the weight vector ($\mathbf{w}$) as a **"template"** or a **"filter"** of what the neuron *wants* to see.

Let's say a neuron has three weights: $\mathbf{w} = [1, 1, -1]$.
This means the neuron *likes* feature 1, *likes* feature 2, and *hates* feature 3.

**Scenario A: The Perfect Match**
An input comes in that exactly matches what the neuron wants: $\mathbf{x} = [1, 1, -1]$.
$$z = (1 \cdot 1) + (1 \cdot 1) + (-1 \cdot -1)$$
$$z = 1 + 1 + 1 = 3$$
Because a negative times a negative is a positive, the alignment score is highly positive. The inputs and weights aligned perfectly.

**Scenario B: The Exact Opposite**
An input comes in that is the exact opposite of what the neuron wants: $\mathbf{x} = [-1, -1, 1]$.
$$z = (1 \cdot -1) + (1 \cdot -1) + (-1 \cdot 1)$$
$$z = -1 - 1 - 1 = -3$$
The score is highly negative. The input misaligned completely with the template.

**Scenario C: A Mixed Bag**
An input comes in that only partially matches: $\mathbf{x} = [1, -1, 0]$.
$$z = (1 \cdot 1) + (1 \cdot -1) + (-1 \cdot 0)$$
$$z = 1 - 1 + 0 = 0$$
The positive and negative matches cancel each other out, resulting in an alignment score of 0.

### Summary
When the neural network updates its weights during training, it is literally rotating and stretching its "weight arrows" in space so that they point in the same direction as the examples it wants to classify as "true," and point away from the examples it wants to classify as "false." The dot product is simply the mathematical tool that measures how well it succeeded.

````python
import torch
import torch.nn as nn

# Define a single hidden layer with 3 inputs and 2 output neurons
class SimpleLayer(nn.Module):
    def __init__(self):
        super().__init__()
        # nn.Linear automatically initializes the W matrix and b vector
        self.linear = nn.Linear(in_features=3, out_features=2)
        self.activation = nn.Sigmoid()

    def forward(self, x):
        z = self.linear(x)        # z = xW + b
        a = self.activation(z)    # a = g(z)
        return a

x_tensor = torch.tensor([1.5, 2.0, -1.0])
model = SimpleLayer()
output = model(x_tensor)
print(f"Layer Output: {output.detach().numpy()}")
````

This is exactly where the math usually trips people up! It is a brilliant question.

When we move to matrix notation, the individual "neurons" don't disappear—they just get packed together into a grid (the matrix) so the computer can calculate them all at the exact same time.

Here is the secret: **In the weight matrix $\mathbf{W}$, every single column represents one individual neuron.**

Let's break down exactly how multiple neurons "come in" and form that matrix.

### 1. The Setup: Two Neurons, One Input
Imagine you have an input vector with 3 features: $\mathbf{x} = [x_1, x_2, x_3]$.

Instead of just one neuron looking at this input, imagine we now have **two separate neurons** (let's call them Neuron A and Neuron B) sitting side-by-side in a layer. Both of them are looking at the exact same input $\mathbf{x}$, but they are trying to learn *different* patterns.

Because they are looking for different things, they have their own personal sets of weights:
* **Neuron A's weights:** $\mathbf{w}_A = [w_{A1}, w_{A2}, w_{A3}]$
* **Neuron B's weights:** $\mathbf{w}_B = [w_{B1}, w_{B2}, w_{B3}]$

If we did the math the old-fashioned way, we would do two separate dot products:
* Neuron A's alignment score: $z_A = (x_1 \cdot w_{A1}) + (x_2 \cdot w_{A2}) + (x_3 \cdot w_{A3})$
* Neuron B's alignment score: $z_B = (x_1 \cdot w_{B1}) + (x_2 \cdot w_{B2}) + (x_3 \cdot w_{B3})$

### 2. Enter the Matrix ($\mathbf{W}$)
Doing separate dot products is slow. To speed this up, we take Neuron A's weight vector and Neuron B's weight vector and "glue" them together to form a matrix.

We stand them up vertically side-by-side. This becomes our weight matrix $\mathbf{W}$:

$$\mathbf{W} = \begin{bmatrix} w_{A1} & w_{B1} \\ w_{A2} & w_{B2} \\ w_{A3} & w_{B3} \end{bmatrix}$$

* **Column 1** is purely Neuron A.
* **Column 2** is purely Neuron B.



### 3. The Matrix Multiplication ($\mathbf{x}\mathbf{W}$)
Now, when we multiply the input vector $\mathbf{x}$ by the entire matrix $\mathbf{W}$, the rules of linear algebra automatically calculate both dot products at the same time:

$$\mathbf{x}\mathbf{W} = [x_1, x_2, x_3] \begin{bmatrix} w_{A1} & w_{B1} \\ w_{A2} & w_{B2} \\ w_{A3} & w_{B3} \end{bmatrix} = [z_A, z_B]$$

The output is a new vector: $[z_A, z_B]$.
* The first number is Neuron A's output.
* The second number is Neuron B's output.

We then add the bias vector $\mathbf{b}$ (which has one bias for Neuron A and one for Neuron B) and pass both numbers through our activation function simultaneously.

### Connecting it back to the Code
When you write this line in PyTorch:
`nn.Linear(in_features=3, out_features=2)`

You are literally telling the computer:
"Create a layer with **2 neurons** (`out_features`). Make sure each of those neurons has **3 weights** (`in_features`) so they can read the input. Pack all of this into a 3x2 matrix."

So, whenever you see a layer in deep learning, just remember that the matrix is just a bookshelf, and every book on that shelf is a single, independent neuron! Does that visual make the transition from vector to matrix clearer?

````python
import torch
import torch.nn as nn

# Define a single hidden layer with 3 inputs and 2 output neurons
class SimpleLayer(nn.Module):
    def __init__(self):
        super().__init__()
        # nn.Linear automatically initializes the W matrix and b vector
        self.linear = nn.Linear(in_features=3, out_features=2)
        self.activation = nn.Sigmoid()

    def forward(self, x):
        z = self.linear(x)        # z = xW + b
        a = self.activation(z)    # a = g(z)
        return a

x_tensor = torch.tensor([1.5, 2.0, -1.0])
model = SimpleLayer()
output = model(x_tensor)
print(f"Layer Output: {output.detach().numpy()}")
````

````python
# True label (e.g., 1.0) and our model's prediction
y_true = torch.tensor([1.0])
y_pred = torch.tensor([0.2]) # The model is very wrong

# Calculate Mean Squared Error
mse_loss_fn = nn.MSELoss()
loss = mse_loss_fn(y_pred, y_true)
print(f"Loss: {loss.item():.4f}")
````

You have hit on one of the biggest conceptual hurdles in deep learning! It is incredibly counterintuitive at first: if the loss is just a single, solitary number (like 2.4), how on earth does the network know how to adjust hundreds, thousands, or billions of individual weights from that one number?

Here is the secret: **Backpropagation doesn't just look at the final number. It looks at the *unbroken mathematical equation* that created that number.**

Let’s break down exactly how this "blame game" works.

### 1. The Forward Pass Leaves a Trail (The Computation Graph)
When you compute the forward pass, the computer isn't just calculating the final loss value; it is secretly building a map of every single math operation that happened along the way. This map is called a **Computation Graph**.



Think of the loss value as the final score at the end of a long relay race. To improve the score, you can't just look at the stopwatch; you have to look at the exact path each runner took to see who slowed the team down.

Because we have the exact equation (the graph), we can use calculus to find the **derivative** (or gradient) of the loss with respect to every single weight.

### 2. What the Gradient Actually Means
A gradient answers one very specific question:
**"If I nudge this specific weight $w$ up by a tiny microscopic amount, will my final Loss number go up or down, and by how much?"**

* If the gradient is a **positive number**, it means increasing the weight *increases* the loss (makes things worse). So, we need to subtract from the weight.
* If the gradient is a **negative number**, it means increasing the weight *decreases* the loss (makes things better). So, we need to add to the weight.
* If the gradient is **zero**, we are at the bottom of the valley. The weight is perfect.

### 3. The Chain Rule: Passing the Blame Backward
To calculate that gradient for a weight deep inside the network, we have to use the **Chain Rule** from calculus. We work backward from the final loss, step-by-step, asking each layer how much it contributed to the error.



Let's look at the math for a single weight $w$ in our simple perceptron. We want to find $\frac{\partial J}{\partial w}$ (the gradient of the Loss $J$ with respect to weight $w$).

We calculate it by multiplying three smaller, easier derivatives together:

$$\frac{\partial J}{\partial w} = \frac{\partial J}{\partial y} \cdot \frac{\partial y}{\partial z} \cdot \frac{\partial z}{\partial w}$$

Here is what each piece means in plain English:
1.  **$\frac{\partial J}{\partial y}$ (Loss to Prediction):** "Hey final prediction ($y$), how much did you mess up the loss?" (Calculated directly from the loss function, like Mean Squared Error).
2.  **$\frac{\partial y}{\partial z}$ (Prediction to Pre-activation):** "Hey Sigmoid function, how sensitive were you to the raw number ($z$) that came into you?" (Calculated from the derivative of the activation function).
3.  **$\frac{\partial z}{\partial w}$ (Pre-activation to Weight):** "Hey weight ($w$), how much did you contribute to that raw number ($z$)?" (Because $z = w \cdot x + b$, the derivative here is simply the input $x$).

By multiplying these three numbers together, the network gets an exact value that says, "Weight 1, you are responsible for 0.4 of this error. Weight 2, you are responsible for -0.1 of this error."

### 4. How it looks in Code
This entire calculus operation is what PyTorch does automatically when you call `loss.backward()`. You don't have to write the Chain Rule by hand; PyTorch traverses the computation graph for you.

```python
import torch

# 1. Setup a simple weight, input, and target
x = torch.tensor([2.0])
y_true = torch.tensor([1.0])

# We tell PyTorch to track the gradients for this specific weight
w = torch.tensor([0.5], requires_grad=True)

# 2. Forward Pass (Building the graph)
z = w * x
y_pred = torch.sigmoid(z)
loss = (y_pred - y_true)**2 # Mean Squared Error

print(f"The single loss value is: {loss.item():.4f}")

# 3. BACKPROPAGATION
# This traverses the graph backward and calculates the Chain Rule for 'w'
loss.backward()

# Now, 'w' has a secret attribute attached to it holding its gradient!
print(f"The gradient for weight 'w' is: {w.grad.item():.4f}")

# 4. Gradient Descent Update
learning_rate = 0.1
with torch.no_grad(): # Don't track this update operation in the graph
    w_new = w - learning_rate * w.grad

print(f"Updated weight: {w_new.item():.4f}")
```

In short, the loss *is* just a number, but because it is tethered to the network via a mathematical equation, we can reverse-engineer exactly how to fix the weights that produced it.

To make this completely concrete, let's walk through a single, complete cycle of learning (Forward Pass $\rightarrow$ Loss $\rightarrow$ Backward Pass $\rightarrow$ Update).

We will use actual numbers, logical steps, and the mathematical formulas to see exactly how a network "thinks" and corrects itself.



### The Setup: Our Tiny Brain
Imagine the simplest possible network: one input ($x$), one weight ($w$), no bias, a Sigmoid activation, and a Mean Squared Error (MSE) loss function.

* **What we see (Input):** $x = 2.0$
* **What we currently believe (Initial Weight):** $w = 0.5$
* **The true answer we *want* to get (Target):** $y_{true} = 1.0$
* **Our Learning Rate ($\eta$):** $0.1$

---

### Step 1: The Forward Pass (Making a Guess)
The network pushes the input through the math to make a prediction.

**1. The Linear Transformation ($z$)**
* **Logic:** We multiply the input by our current weight.
* **Formula:** $z = w \cdot x$
* **Calculation:** $z = 0.5 \cdot 2.0 = 1.0$

**2. The Activation Function ($y_{pred}$)**
* **Logic:** We squash $z$ into a probability between 0 and 1 using Sigmoid.
* **Formula:** $y_{pred} = \frac{1}{1 + e^{-z}}$
* **Calculation:** $y_{pred} = \frac{1}{1 + e^{-1.0}} \approx 0.73$

**3. The Loss ($J$)**
* **Logic:** We measure how wrong our prediction of 0.73 is compared to the true answer of 1.0 using Mean Squared Error.
* **Formula:** $J = (y_{pred} - y_{true})^2$
* **Calculation:** $J = (0.73 - 1.0)^2 = (-0.27)^2 \approx 0.073$

**The Reality Check:** Our loss is $0.073$. We predicted $0.73$, but we wanted $1.0$. Our prediction is **too low**.

---

### Step 2: The Backward Pass (The Chain Rule)
Now, we must figure out how to change $w$ to reduce that $0.073$ error. We work backward from the error to the weight, calculating three separate "local gradients" using the Chain Rule:

$$\frac{\partial J}{\partial w} = \frac{\partial J}{\partial y_{pred}} \cdot \frac{\partial y_{pred}}{\partial z} \cdot \frac{\partial z}{\partial w}$$

Let's calculate each piece:

**Part A: The Loss Gradient ($\frac{\partial J}{\partial y_{pred}}$)**
* **Logic:** "If my prediction goes up, does my error go up or down?"
* **Formula:** The derivative of $(y_{pred} - y_{true})^2$ is $2(y_{pred} - y_{true})$.
* **Calculation:** $2(0.73 - 1.0) = -0.54$
* *Meaning:* Because this is a negative number, it tells us: "If you increase $y_{pred}$, the error will go down."

**Part B: The Activation Gradient ($\frac{\partial y_{pred}}{\partial z}$)**
* **Logic:** "If the raw number $z$ goes up, how much does the Sigmoid output change?"
* **Formula:** The derivative of the Sigmoid function is incredibly elegant: $y_{pred} \cdot (1 - y_{pred})$.
* **Calculation:** $0.73 \cdot (1 - 0.73) = 0.73 \cdot 0.27 \approx 0.197$
* *Meaning:* A change in $z$ transfers through the Sigmoid at a rate of roughly 20%.

**Part C: The Weight Gradient ($\frac{\partial z}{\partial w}$)**
* **Logic:** "If I change the weight $w$, how much does $z$ change?"
* **Formula:** Since $z = w \cdot x$, the derivative with respect to $w$ is simply $x$.
* **Calculation:** $x = 2.0$
* *Meaning:* The input acts as an amplifier. Because the input is $2.0$, any change we make to the weight is doubled when calculating $z$.



**Putting it all together (The Final Gradient):**
We multiply the three pieces together to get our final gradient for the weight:
$$\frac{\partial J}{\partial w} = (-0.54) \cdot (0.197) \cdot (2.0) \approx -0.213$$

**The Grand Conclusion of the Backward Pass:**
The final gradient is **-0.213**. The negative sign is crucial! It tells the network: "Because the gradient is negative, increasing the weight $w$ will decrease the total loss."

---

### Step 3: Gradient Descent (The Weight Update)
Now we apply the fix. We update our old weight using the gradient and our learning rate ($\eta = 0.1$).

* **Logic:** We take the current weight and step in the *opposite* direction of the gradient to find the minimum error.
* **Formula:** $w_{new} = w - \eta \cdot \frac{\partial J}{\partial w}$
* **Calculation:** $w_{new} = 0.5 - (0.1 \cdot -0.213)$
* $w_{new} = 0.5 - (-0.0213)$
* $w_{new} = 0.5213$

### Did it work? (The Proof)
Our weight increased from **0.5** to **0.5213**.
Let's fast-forward and see what happens the next time the network sees the same input ($x = 2.0$):
* New $z = 0.5213 \cdot 2.0 = 1.0426$
* New $y_{pred} = \frac{1}{1 + e^{-1.0426}} \approx \textbf{0.74}$

**Success!** Our prediction went up from $0.73$ to $0.74$, getting closer to our target of $1.0$. The loss decreased. The network just successfully learned! By repeating this loop thousands of times, the weight will eventually perfectly align to predict $1.0$.

Instead, talking about a "1 unit change" is just the universal way mathematicians explain what a **derivative** (or slope) is. Let me explain why we use that phrasing and what it actually means for the neural network.

---

### The Concept of a Derivative (The "Per 1 Unit" Rule)

A derivative is just a measure of **speed** or **sensitivity**. It measures a ratio of how two things move together.

Think about driving a car. If your speedometer says **60 miles per hour**, what does that mean?
It means: *"If I drive for **1 hour**, my distance will change by 60 miles."* You don't *have* to drive for exactly 1 hour. You might only drive for 5 minutes! But we still describe your speed using that "per 1 unit of time" ratio.



In calculus, a derivative $\frac{\partial y}{\partial z}$ is the exact same thing. It is the "speed" at which $y$ changes when $z$ changes.
When we calculate that the derivative of our Sigmoid function is **0.197**, we are calculating a speed.

It means: *"At this exact moment, the output $y_{pred}$ is moving at a speed of 0.197 units per **1 unit** of $z$."*

### Applying it to the Neural Network (Tiny Nudges)

Here is where we bring it back to reality. In a neural network, we almost never change $z$ by a massive whole number like 1.0. We make tiny, microscopic changes.

But because we know the "speed" (the derivative), we can predict exactly what those tiny changes will do!

Let's say our network takes a tiny optimization step and increases $z$ by just **0.01**. How much will our prediction ($y_{pred}$) change?

We just multiply the tiny change by our derivative (our speed):
$$\text{Change in } y_{pred} = (\text{Change in } z) \times (\text{Derivative})$$
$$\text{Change in } y_{pred} = (0.01) \times (0.197) = 0.00197$$

So, our prediction would go up by roughly $0.00197$.

### Why this is the whole point of Backpropagation

This is why we calculate derivatives in the first place! The neural network is asking:
*"I need to decrease my loss. If I nudge this weight by a tiny 0.01 amount, what will happen at the very end of the line?"*

By calculating the derivative at every step (the weight, the Sigmoid, the loss), the network gets a clear ratio. It knows exactly how sensitive the final error is to that specific weight, allowing it to adjust the weight precisely enough to improve the score on the next try.

Does making that distinction between the "target label" and the "speed ratio" clear up why we talk about it that way?

````python
import torch.optim as optim

# 1. Setup Model, Data, Loss, and Optimizer
model = nn.Linear(3, 1) # 3 inputs, 1 output
optimizer = optim.SGD(model.parameters(), lr=0.01) # eta = 0.01
loss_fn = nn.MSELoss()

x_train = torch.tensor([1.5, 2.0, -1.0])
y_train = torch.tensor([1.0])

# --- THE TRAINING LOOP ---
epochs = 100
for epoch in range(epochs):
    # Step 1: Forward Pass
    y_pred = model(x_train)

    # Step 2: Compute Loss
    loss = loss_fn(y_pred, y_train)

    # Step 3: Zero gradients (clear old gradients)
    optimizer.zero_grad()

    # Step 4: Backpropagation (Compute gradients using the chain rule)
    loss.backward()

    # Step 5: Gradient Descent Step (W_new = W_old - lr * gradient)
    optimizer.step()

    if epoch % 20 == 0:
        print(f"Epoch {epoch} | Loss: {loss.item():.4f}")
````

### 1. The Single Perceptron (The Forward Pass)
The perceptron is the fundamental computational unit. It takes an input vector, scales it by a weight vector, applies a bias shift, and passes the result through a non-linear activation function.

**The Mathematics:**
Given an input vector $\mathbf{x} = [x_1, x_2, ..., x_m]$ and a weight vector $\mathbf{w} = [w_1, w_2, ..., w_m]$:
1.  **Linear Combination:** We compute the dot product of the inputs and weights, plus a scalar bias $b$.
    $$z = \sum_{i=1}^{m} x_i w_i + b = \mathbf{w}^T\mathbf{x} + b$$
2.  **Activation:** We pass $z$ through a non-linear function $g(z)$. A common choice for teaching is the Sigmoid function, which squishes outputs between 0 and 1:
    $$g(z) = \frac{1}{1 + e^{-z}}$$
    $$y = g(z)$$



**The Code (Python/NumPy):**
```python
import numpy as np

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# 1. Initialize inputs, weights, and bias
x = np.array([1.5, 2.0, -1.0]) # Example input features
w = np.array([0.5, -0.2, 0.1]) # Learned weights
b = 0.1                        # Learned bias

# 2. Linear combination (Dot product)
z = np.dot(w, x) + b

# 3. Activation
y = sigmoid(z)
print(f"Neuron Output: {y:.4f}")
```

### 2. Multi-Layer Networks (Matrix Notation)
To process complex representations, we stack multiple neurons together into a layer, and stack layers to form a network. Instead of computing individual dot products, we use matrix multiplication to compute the forward pass for an entire layer at once.

**The Mathematics:**
If a layer has $n$ neurons and receives an input vector $\mathbf{x}$ of size $m$:
* The weights become a matrix $\mathbf{W}$ of shape $(m, n)$.
* The biases become a vector $\mathbf{b}$ of size $n$.
The entire layer is computed as:
$$\mathbf{Z} = \mathbf{x} \mathbf{W} + \mathbf{b}$$
$$\mathbf{A} = g(\mathbf{Z})$$



**The Code (PyTorch):**
PyTorch handles the matrix abstractions internally using `nn.Linear`.

```python
import torch
import torch.nn as nn

# Define a single hidden layer with 3 inputs and 2 output neurons
class SimpleLayer(nn.Module):
    def __init__(self):
        super().__init__()
        # nn.Linear automatically initializes the W matrix and b vector
        self.linear = nn.Linear(in_features=3, out_features=2)
        self.activation = nn.Sigmoid()

    def forward(self, x):
        z = self.linear(x)        # z = xW + b
        a = self.activation(z)    # a = g(z)
        return a

x_tensor = torch.tensor([1.5, 2.0, -1.0])
model = SimpleLayer()
output = model(x_tensor)
print(f"Layer Output: {output.detach().numpy()}")
```

### 3. Quantifying Error (The Loss Function)
To teach a network, we must mathematically define how "wrong" its current predictions are compared to the ground truth. This is the Empirical Risk or Loss Function $J$.

**The Mathematics:**
For a continuous output (regression), we commonly use **Mean Squared Error (MSE)** over $N$ samples:
$$J(\mathbf{W}) = \frac{1}{N} \sum_{i=1}^{N} (y_i - \hat{y}_i)^2$$
Where $y_i$ is the true label and $\hat{y}_i$ is our model's prediction.

**The Code:**
```python
# True label (e.g., 1.0) and our model's prediction
y_true = torch.tensor([1.0])
y_pred = torch.tensor([0.2]) # The model is very wrong

# Calculate Mean Squared Error
mse_loss_fn = nn.MSELoss()
loss = mse_loss_fn(y_pred, y_true)
print(f"Loss: {loss.item():.4f}")
```

### 4. Backpropagation and Gradient Descent
This is the core engine of deep learning. We need to find how a tiny change in a specific weight $w_{ij}$ affects the total loss $J$. We do this by calculating the gradient (the multidimensional derivative) of the loss with respect to the weights.

**The Mathematics:**
By applying the **Chain Rule** from calculus, we work backward from the loss to the weights:
$$\frac{\partial J}{\partial w} = \frac{\partial J}{\partial \hat{y}} \cdot \frac{\partial \hat{y}}{\partial z} \cdot \frac{\partial z}{\partial w}$$

Once we have the gradient $\nabla J(\mathbf{W})$, we update the weights in the *opposite* direction of the gradient to minimize the loss. We scale the step by the **learning rate** $\eta$:
$$\mathbf{W}_{new} = \mathbf{W}_{old} - \eta \nabla J(\mathbf{W})$$



**The Code (The Full Training Loop):**
PyTorch's Autograd engine computes the chain rule automatically when we call `.backward()`.

```python
import torch.optim as optim

# 1. Setup Model, Data, Loss, and Optimizer
model = nn.Linear(3, 1) # 3 inputs, 1 output
optimizer = optim.SGD(model.parameters(), lr=0.01) # eta = 0.01
loss_fn = nn.MSELoss()

x_train = torch.tensor([1.5, 2.0, -1.0])
y_train = torch.tensor([1.0])

# --- THE TRAINING LOOP ---
epochs = 100
for epoch in range(epochs):
    # Step 1: Forward Pass
    y_pred = model(x_train)
    
    # Step 2: Compute Loss
    loss = loss_fn(y_pred, y_train)
    
    # Step 3: Zero gradients (clear old gradients)
    optimizer.zero_grad()
    
    # Step 4: Backpropagation (Compute gradients using the chain rule)
    loss.backward()
    
    # Step 5: Gradient Descent Step (W_new = W_old - lr * gradient)
    optimizer.step()
    
    if epoch % 20 == 0:
        print(f"Epoch {epoch} | Loss: {loss.item():.4f}")
```

### Teaching Tips
When presenting this, start with the raw NumPy perceptron. It proves that there is no "magic" happening—it's just arithmetic. Then, introduce the PyTorch equivalents to show how modern frameworks abstract the calculus (`loss.backward()`) and matrix tracking, allowing researchers to scale from a single layer to billions of parameters without rewriting the foundational math.
